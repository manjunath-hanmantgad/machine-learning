```python
# grouping rows according to some value

import pandas as pd
url ='https://raw.githubusercontent.com/chrisalbon/simulated_datasets/master/titanic.csv' # create url 
dataframe = pd.read_csv(url) # load data as dataframe

dataframe.groupby('Sex').mean()
```

```python
dataframe.groupby('Survived')['Name'].count()

# 0 is for male and 1 is for female
```

```python
dataframe.groupby(['Sex','Survived'])['Age'].mean()
```

```python
dataframe.groupby(['Sex','Survived'])['Age'].sum()
```

```python
dataframe.groupby(['Sex','Survived'])['Age'].count()

# so here female survived = 217 and male survived = 96 ! :( so unfair!!
```

```python
# Looping inside 

for name in dataframe['Name'][:10]:
    print(name.upper())
    
# this will print first 10 names with all capital letters.
```

```python
# same using comprehensions

[name.upper() for name in dataframe['Name'][:10]]
```

```python
[name.lower() for name in dataframe['Name'][:10]]
```

```python
# same functionality using functions.

def uppercase(x):
    return x.upper()

# apply this function upper to dataframe

dataframe['Name'].apply(uppercase)[:5]
```

```python
def lowercase(x):
    return x.lower()
dataframe['Name'].apply(lowercase)[:5]
```

```python
# applying groupby and functions to dataframe

dataframe.groupby('Sex').apply(lambda x: x.count())
```

```python
# merging data frames

import pandas as pd

# dataframe 1

employee_data = {'employee_id':[1,2,3,4],
                'name': ['Tony','Thor','Steve','Natasha']}
dataframe_employees = pd.DataFrame(employee_data , columns = ['employee_id', 'name'])

# dataframe 2

sales_data = {'employee_id':[2,3,4,5],
             'total_sales': [1000,2000,3000,4000]}
dataframe_sales = pd.DataFrame(sales_data,columns = ['employee_id', 'total_sales'])


# now merge 

# merge on employee id

pd.merge(dataframe_employees , dataframe_sales , on = 'employee_id')
```

```python
# now outer join

pd.merge(dataframe_employees , dataframe_sales , on = 'employee_id' , how = 'outer')

# outer returns all rows from both datframes , if a row exist in one df and not in other df then taht value is replaced using NaN.
```

```python
pd.merge(dataframe_employees , dataframe_sales , on = 'employee_id' , how = 'inner') # this is default
```

```python
pd.merge(dataframe_employees , dataframe_sales , on = 'employee_id' , how = 'left')
```

```python
pd.merge(dataframe_employees , dataframe_sales , on = 'employee_id' , how = 'right') # right will start from 2 from df 2
```
